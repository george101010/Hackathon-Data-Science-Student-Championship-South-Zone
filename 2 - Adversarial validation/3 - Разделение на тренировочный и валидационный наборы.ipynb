```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import  kstest  
from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

Проверка равенства распределений при разделеннии набора train на train и validation

Их соотношение:
* train -  80%
* validation 20%

Загрузка данных

```python
train = pd.read_csv('train.csv' , sep = ',').drop([ 'Property_ID','locality' ] , axis = 1)
test = pd.read_csv('test.csv' , sep = ',').drop([ 'Property_ID','locality' ] , axis = 1)
```

```python
cat_cols = train.columns[ (train.columns != 'security_deposit') & (train.columns != 'area' )&  (train.columns != 'price' )  ]
```

```python
train[ cat_cols ] =  train[ cat_cols ].astype("category")
test[cat_cols ] =  test[ cat_cols].astype("category")
```

```python
# логарифмирование целевой переменной
Y = np.log1p( train['price'])
train.drop([ 'price' ],axis=1 , inplace = True)
test.drop([ 'price' ],axis=1 , inplace = True)
```

```python
# разделени на train  и validation
x_train , x_val , y_train , y_val = train_test_split(train , Y , random_state = 67 , test_size = 0.2 )
```

```python
for I in train.columns:
    print( 'Проверка распределения для признака ', I)
    stat , pv =  kstest(rvs = x_train[ I ] , 
                        cdf = x_val[ I ])
    if pv < 0.05:
        print('Pval = ',pv  ,' Распределения неодинаковы')
    else:
        print('Pval = ',pv  ,' Распределения одинаковы')
```

```python
# проверка для целевой переменной
stat , pv =  kstest(rvs = y_train , 
                        cdf = y_val)
if pv < 0.05:
    print('Pval = ',pv  ,' Распределения неодинаковы')
else:
    print('Pval = ',pv  ,' Распределения одинаковы')
```

Разделение на train и validation прошло успешно. Распределения признаков одинаковы.

Проверка равенства распределений для парных взаимодействий

```python
two_way_inter_train = pd.read_csv('two_way_inter_train.csv' , sep=','  )
two_way_inter_test = pd.read_csv('two_way_inter_test.csv' , sep=','  )
```

```python
x_2_way_train , x_2_way_val , _ , _ = train_test_split(two_way_inter_train , Y , random_state = 67 , test_size = 0.2 )
```

Колонки распределения , которых совпали запишутся в equal_distr_2_way , а которых не совпали в nonequal_distr_2_way

```python
equal_distr_2_way = []
nonequal_distr_2_way = []
for I in two_way_inter_train.columns:
    print( 'Проверка распределения для признака ', I)
    stat , pv =  kstest(rvs = x_2_way_train[ I ] , 
                        cdf = x_2_way_val[ I ])
    if pv < 0.05:
        print('Pval = ',pv  ,' Распределения неодинаковы')
        nonequal_distr_2_way.append( I )
    else:
        print('Pval = ',pv  ,' Распределения одинаковы')
        equal_distr_2_way.append( I )
```

```python
 len(equal_distr_2_way) == len(two_way_inter_train.columns)
```

```python
nonequal_distr_2_way
```

```python
two_way_inter_train[equal_distr_2_way].to_csv('two_way_inter_train_ED.csv' , sep=',' , index = False)
two_way_inter_test[equal_distr_2_way].to_csv('two_way_inter_test_ED.csv' , sep=',' , index = False)
```
