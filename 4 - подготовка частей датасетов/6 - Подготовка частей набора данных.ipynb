```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
from scipy.stats import  kstest  
```

Загрузка чатей датасета (Основная часть и взаимодействия)

```python
train = pd.read_csv('train.csv' , sep = ',') 
test = pd.read_csv('test.csv' , sep = ',')
train2w = pd.read_csv('two_way_inter_train_ED_check.csv' , sep = ',')
 
train3w = pd.read_csv('three_way_inter_train_ED_check.csv' , sep = ',')
 
```

Исправление названий столбцов и значений в них. Приведение к одинаковому виду в TRAIN и TEST.

```python
train2w.columns = [ I.replace('.x.' , ' x ') for I in train2w.columns ]
train2w.columns = [ I.replace('geyser solar' , 'geyser/solar') for I in train2w.columns ]
train2w.columns = [ I.replace('CCTV security' , 'CCTV/security') for I in train2w.columns ]
```

```python
train3w.columns = [ I.replace('.x.' , ' x ') for I in train3w.columns ]
train3w.columns = [ I.replace('geyser.solar' , 'geyser/solar') for I in train3w.columns ]
train3w.columns = [ I.replace('CCTV.security' , 'CCTV/security') for I in train3w.columns ]
```

```python
test2w = pd.DataFrame()
for i in tqdm( train2w.columns ) :
    
    coln = test[ i.split(' x ')].apply( lambda x: ' x '.join( str(u) for u in x), axis = 1)
    coln.name = i
    test2w = pd.concat([test2w , coln] , axis =1)
```

```python
test3w = pd.DataFrame()
for i in tqdm( train3w.columns ) :
    
    coln = test[ i.split(' x ')].apply( lambda x: ' x '.join( str(u) for u in x), axis = 1)
    coln.name = i
    test3w = pd.concat([test3w , coln] , axis =1)
```

```python
test2w.to_csv('two_way_inter_test.csv',sep = ',' , index = False)
test3w.to_csv('three_way_inter_test.csv',sep = ',' , index = False)
```

```python
w4_cols = pd.read_csv('fourw_way_inter.csv' , sep = ',' , encoding='ISO-8859-1')
```

```python
cols_4_train = pd.DataFrame()
cols_4_test = pd.DataFrame()

for I in tqdm( w4_cols['4_way_inter'].values ):
    print( I.replace('.','/').split(' è ') )
    tr = train[I.replace('.','/').split(' è ') ].apply( lambda x: ' x '.join( str(u) for u in x), axis = 1)
    tr.name = I.replace('.','/').replace(' è ',' x ')
    
    ts = test[ I.replace('.','/').split(' è ') ].apply( lambda x: ' x '.join( str(u) for u in x), axis = 1)
    ts.name = I.replace('.','/').replace(' è ',' x ')  
    
    stat , pv =  kstest(rvs = tr , 
                        cdf = ts )
    if pv < 0.05:
        print('Pval = ',pv  ,' Распределения неодинаковы')
    else:
        cols_4_train = pd.concat([ cols_4_train ,tr ] , axis=1)
        cols_4_test = pd.concat([ cols_4_test ,ts ] , axis=1)
        print('Pval = ',pv  ,' Распределения одинаковы')
    
```
